```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
# Series.plot.hist() = HISTOGRAM PLOT
fig, ax = plt.subplots()
ax = tips['total_bill'].plot.hist()
plt.show()
```

```python
# Series.plot.hist(alpha=0.5) = HISTOGRAM PLOT times 2
fig, ax = plt.subplots()
ax = tips[['total_bill', 'tip']].plot.hist(alpha=0.5, bins=20, ax=ax) # 0.5 Alpha to show the overlapped histograms plots
plt.show()
```

```python
# Series.plot.kde = Kernel Density Plot
fig, ax = plt.subplots()
ax = tips['tip'].plot.kde()
plt.show()
```

```python
# DataFrame.plot.scatter() = scatterplot
fig, ax = plt.subplots()
ax = tips.plot.scatter(x='total_bill', y='tip', ax=ax)
plt.show()
```

```python
# DataFrame.plot.hexbin() = Hexbin Plot
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x='total_bill', y='tip', gridsize=10, ax=ax)
plt.show()
```

```python
# DataFrame.plot.box() = Box Plot
fig, ax = plt.subplots()
ax = tips.plot.box(ax=ax)
plt.show()
```
