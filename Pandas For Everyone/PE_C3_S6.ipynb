```python
import matplotlib.pyplot as plt
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
# Color Schemes, comparison
fig, ax = plt.subplots()
ax = sns.violinplot(x='time', y='total_bill', 
                   hue='sex', data=tips, 
                   split=True)
plt.show()

sns.set_style('whitegrid')
ax = sns.violinplot(x='time', y='total_bill', 
                   hue='sex', data=tips, 
                   split=True)
plt.show()
```

```python
# This is extremely complicated.
fig = plt.figure()
seabornStyles = ['whitegrid', 'darkgrid', 'ticks', 'white', 'dark']
for idx, style in enumerate(seabornStyles):
    plotPos = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plotPos)
        violin = sns.violinplot(x='time', y='total_bill',
                               data=tips, ax=ax)
        violin.set_title('style')
fig.tight_layout()
plt.show()
```
