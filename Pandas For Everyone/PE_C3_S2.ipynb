```python
import seaborn as sns # Importing seaborn bpackage
anscombe = sns.load_dataset("anscombe") # anscombe data, used in statistics to show off graphs
print(type(anscombe)) # Just want to check to see what it is, Pandas DataFrame!
```

```python
import matplotlib.pyplot as plt # Plot time
dataset1 = anscombe[anscombe["dataset"] == "I"] # subsetting the data
```

```python
plt.plot(dataset1['x'], dataset1['y'])
plt.show() # This comes out crazy, I guess the package is intepretting the plot since it isn't explicitly indicated.
```

```python
plt.plot(dataset1['x'], dataset1['y'], 'o') # Here we go, I guess this will be a scatterplot
plt.show() # Scatter plot!
```

```python
dataset2 = anscombe[anscombe["dataset"] == 'II'] # Same process with the other categories in the the dataset column
dataset3 = anscombe[anscombe["dataset"] == 'III']
dataset4 = anscombe[anscombe["dataset"] == 'IV']
```

```python
plt.plot(dataset1['x'], dataset1['y'], 'o')
plt.plot(dataset2['x'], dataset2['y'], 'x')
plt.plot(dataset3['x'], dataset3['y'], '+')
plt.plot(dataset4['x'], dataset4['y'], 's')
plt.show() # I did not know this would all be plotted at the same time! NICE!
```

```python
# We can plot everything at the same time, in separate plots using figure()
fig = plt.figure()
axesI = fig.add_subplot(2, 2, 1)
axesII = fig.add_subplot(2, 2, 2)
axesIII = fig.add_subplot(2, 2, 3)
axesIV = fig.add_subplot(2, 2, 4) # I love this!

axesI.plot(dataset1['x'], dataset1['y'], 'o')
axesII.plot(dataset2['x'], dataset2['y'], 'o')
axesIII.plot(dataset3['x'], dataset3['y'], 'o')
axesIV.plot(dataset4['x'], dataset4['y'], 'o') # Here we go!

axesI.set_title('I')
axesII.set_title('II')
axesIII.set_title('III')
axesIV.set_title('IV')

fig.suptitle("Anscombe Data")
fig.tight_layout()
```
